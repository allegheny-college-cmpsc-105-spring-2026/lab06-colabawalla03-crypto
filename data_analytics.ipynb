```python
# Analytics

# : Do one or more explorations with dfa
# no interaction is required

# make a blank canvas

# make a title starting with 'Figure 1:' Fill in a relevant title
plt.title('')

# plot
# ylabel
# xlabel
# ylim
# etc.

# reveal everything

import pandas as pd
import matplotlib.pyplot as plt

dfa_city = pd.read_csv(
    '/content/data_analytics.csv',
    sep=',',
    skiprows=74,
    nrows=5
)

print(dfa_city)

plt.figure(figsize=(8, 5))
plt.bar(dfa_city['City'], dfa_city['Active users'])
plt.title('Figure 1: Active Users by City')
plt.xlabel('City')
plt.ylabel('Active users')
plt.ylim(0, dfa_city['Active users'].max() + 1)
plt.tight_layout()
plt.show()
```

```python
# Survey

# : Do one or more explorations with dfs
# no interaction is required

# make a blank canvas

# make a title starting with 'Figure 2:' Fill in a relevant title
plt.title('')

# plot
# ylabel
# xlabel
# ylim
# etc.

# reveal everything

import pandas as pd
import matplotlib.pyplot as plt

# Load the survey data
dfs = pd.read_csv('/content/data_survey.csv')

# Count preferences
pet_counts = dfs['What do you prefer'].value_counts()

# make a blank canvas
plt.figure(figsize=(8, 5))

# plot
plt.bar(pet_counts.index, pet_counts.values)

# title
plt.title('Figure 2: Dog vs Cat Preferences')

# labels
plt.xlabel('Preference')
plt.ylabel('Number of responses')

# ylim
plt.ylim(0, pet_counts.max() + 1)

plt.tight_layout()
plt.show()
```
